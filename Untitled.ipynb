```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
dataset=fetch_california_housing()
```

```python
dataset
```

```python
x=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
y=pd.DataFrame(dataset.target,columns=dataset.target_names)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=21)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
cls=DecisionTreeRegressor()
```

```python
cls.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(10,10)
tree.plot_tree(cls,filled=True);
```

```python
y_pred=cls.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_pred,y_test)
```

```python
parameter={
    "criterion":['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    "splitter":['best', 'random'],
    "max_depth":list(range(13)),
    'max_features':['auto', 'sqrt', 'log2']
}
```

```python
regressor=DecisionTreeRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressorcv=GridSearchCV(regressor,param_grid=parameter,cv=2)
```

```python
reg=regressorcv.fit(x_train,y_train)
```
