```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
dataset
```

```python
y=dataset.target
```

```python
y
```

```python
x=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
#no feature engineing 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=22,test_size=.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
cls=DecisionTreeClassifier(criterion='entropy')
```

```python
cls.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,10))
tree.plot_tree(cls,filled=True);
```

```python
x
```

# Post Pruning

```python
classs=DecisionTreeClassifier(criterion='entropy',max_depth=2)
```

```python
classs.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,10))
tree.plot_tree(classs,filled=True);
```

```python
y_pred=classs.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score , classification_report
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# DecisionTree Prepruning and hyperparameter tuning for huge data

```python
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier=DecisionTreeClassifier()
clf=GridSearchCV(classifier,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
clf.fit(x_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score , classification_report,confusion_matrix
```
